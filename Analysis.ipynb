### About Column names


invoice = Invoice number. Nominal. A 6-digit integral number uniquely assigned to each transaction. If this code starts with the letter 'c', it indicates a cancellation.

StockCode = Product (item) code. Nominal. A 5-digit integral number uniquely assigned to each distinct product.

Description = Product (item) name. Nominal.

Quantity = The quantities of each product (item) per transaction. Numeric.

InvoiceDate = Invice date and time. Numeric. The day and time when a transaction was generated.

Price = Unit price. Numeric. Product price per unit in .

Customer ID = Customer number. Nominal. A 5-digit integral number uniquely assigned to each customer

Country = Country name. Nominal. The name of the country where a customer resides.

### procedure of fill null value in Customer ID column

In above data after observing carefully we see that Invoice and Customer ID column are related.Each value of Customer ID column is same as Invoice values. Acording to Customer ID column they have 5-digit integral number uniquely assigned to each customer.So in the null values of Customer ID column, we are create new unique customer ID in the basis of Invoice Column.Hence the Procedure is given below

1)After fillng null values by '0' we are split null and non null dataset in data0 and data.
2)Create seprate column of unique Invoice column values(af)
3)Create 5 digit unique values(m) and combine both 'm' and 'af' in dataset1
4)merge dataset1 into data
5)After some simplification create dataset6 has to new non null value dataset

```python
dataset['Customer ID'].fillna(0,inplace=True)
```

```python
data0=dataset.loc[dataset['Customer ID']!=0]
data0.head()
```

```python
data=dataset.loc[dataset['Customer ID']==0]
data.head()
```

```python
abc=data['Invoice'].unique()
af = pd.DataFrame(abc, columns= ['Invoice'])
af.head()
```

```python
a=range(20000,23593,1)
m=pd.DataFrame(a,columns=['ID'])
m.head()
```

```python
dataset1=pd.concat([af,m],axis=1)
dataset1.head()
```

```python
dataset2=data.merge(dataset1,on='Invoice',how='left')
dataset2.head()
```

```python
dataset3=dataset2.drop(columns="Customer ID")
dataset4=dataset3.iloc[:,[0,1,2,3,4,5,7,6]]
dataset5=dataset4.rename(columns={"ID":"Customer ID"})
dataset5.head()
```

```python
dataset6=pd.concat([data0,dataset5],ignore_index=True)
```
